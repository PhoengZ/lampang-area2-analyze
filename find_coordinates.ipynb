```python
import pandas as pd
from geopy.geocoders import ArcGIS
import time
```

```python
file_path = 'unit_position.csv'
df = pd.read_csv(file_path)
```

```python
geolocator = ArcGIS()
```

```python
def get_coordinates(row):
    place = str(row['place']).strip()
    village = str(row['village']).strip()
    sub_district = str(row['sub-district']).strip()
    district = str(row['district']).strip()
    province = str(row['province']).strip()

    address = f"{place} หมู่ {village} ต.{sub_district} อ.{district} จ.{province}"

    try:
        # ลองหาแบบละเอียด
        location = geolocator.geocode(address)

        if location:
            print(f"✅ พบ: {place}")
            return pd.Series([location.latitude, location.longitude])
        else:
            # Fallback 1: ตัดชื่อสถานที่ออก ลองหาแค่ หมู่บ้าน + ตำบล + อำเภอ
            fallback_1 = f"หมู่ {village} ต.{sub_district} อ.{district} จ.{province}"
            loc_fb1 = geolocator.geocode(fallback_1)

            if loc_fb1:
                print(f"⚠️ พบแค่ระดับหมู่บ้าน -> ใช้พิกัด หมู่ {village}")
                return pd.Series([loc_fb1.latitude, loc_fb1.longitude])

            # Fallback 2: หาแค่ระดับตำบล
            fallback_2 = f"ต.{sub_district} อ.{district} จ.{province}"
            loc_fb2 = geolocator.geocode(fallback_2)

            if loc_fb2:
                print(f"⚠️ พบแค่ระดับตำบล -> ใช้พิกัด ต.{sub_district}")
                return pd.Series([loc_fb2.latitude, loc_fb2.longitude])

            print(f"❌ ค้นหาไม่พบเลยสำหรับ: {address}")
            return pd.Series([None, None])

    except Exception as e:
        print(f"Error สำหรับ {address}: {e}")
        time.sleep(1)
        return pd.Series([None, None])
```

```python
df[['latitude', 'longitude']] = df.apply(get_coordinates, axis=1)
```

```python
output_file = 'output_with_coordinates2.csv'
df.to_csv(output_file, index=False, encoding='utf-8-sig')
```
